```python
sales = float(input("Enter Sales ($): "))
rate = float(input("Enter Tax Rate (%): "))

cogs = sales * 0.4
gross_profit = sales - cogs
taxes = rate * gross_profit
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator: ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation" )
  res = None

print(f"Result: {res}")
```
